```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

# Prompt the user to enter financial details
income = float(input("Enter your income: $ "))  # User's income
risk_level = input("Enter your risk level (Low, Medium, High): ")  # Risk preference
start_date = input("Enter the start date (YYYY-MM-DD): ")  # Analysis start date
end_date = input("Enter the end date (YYYY-MM-DD): ")  # Analysis end date

# Function to fetch a predefined list of popular stock tickers from Yahoo Finance
def get_all_stocks():
    # Expanded list of 500 stock tickers
    return [
        "AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "META", "NFLX", "NVDA", "SPY", "QQQ",
        "IBM", "WMT", "GE", "DIS", "BA", "V", "JNJ", "PFE", "KO", "XOM", "CVX", "INTC",
        "T", "CSCO", "ORCL", "BABA", "NKE", "AMD", "SBUX", "GS", "HD", "MCD", "LULU",
        "PYPL", "UBER", "SQ", "WFC", "CRM", "SHOP", "TSM", "SNAP", "BA", "LMT", "CAT",
        "RTX", "ZION", "AMAT", "INTU", "MU", "ATVI", "RCL", "UAL", "F", "GE", "VZ", "MS",
        "MDT", "CVX", "MCK", "TMO", "BNS", "C", "AIG", "BMY", "MA", "GS", "FISV", "TGT",
        "PFE", "HPE", "LNC", "T", "DHR", "PFE", "NVS", "AZN", "SNE", "HLT", "VEEV", "BA",
        "AIG", "CI", "HUM", "AMT", "VZ", "LLY", "ABBV", "VLO", "PSX", "SYY", "PG", "CLX",
        "PEP", "KO", "CVS", "WBA", "MCK", "MSFT", "GOOG", "PFE", "MGM", "UAL", "RTX", "EXC",
        "ABBV", "AAPL", "NVDA", "AMD", "FB", "DIS", "LULU", "ACN", "NKE", "INTC", "GS", "MS",
        "COF", "STZ", "UNH", "MDT", "XOM", "ETSY", "FDX", "INTU", "TMO", "PFG", "APD", "EXR",
        "SBUX", "ABT", "JNJ", "STZ", "NXPI", "CL", "WMT", "PBA", "NXST", "LNT", "TSCO", "EMR",
        "SWKS", "WDC", "WFC", "AMGN", "BLK", "V", "ITW", "HCA", "AON", "NEE", "DHR", "TMO",
        "CHTR", "FIS", "WBA", "MU", "AAPL", "TGT", "COST", "AMZN", "BABA", "HD", "TRV", "MRK",
        "VLO", "COP", "SIVB", "GM", "F", "X", "GM", "CVX", "CVS", "FISV", "SYY", "NEE",
        "INTC", "BA", "GS", "SBUX", "MCD", "WMT", "PYPL", "BABA", "TSLA", "GOOG", "T", "AMZN",
        "AMAT", "MA", "SNE", "TMO", "C", "PYPL", "BA", "V", "VZ", "EBAY", "ACN", "CSCO",
        "BBY", "LULU", "WFC", "ETSY", "TGT", "ORCL", "CVX", "PFE", "NKE", "GE", "BMY", "KO",
        "ADBE", "SPLK", "AIG", "AMZN", "EXC", "TMO", "MRK", "NVDA", "BA", "SYY", "JPM", "IBM",
        "AAPL", "AMD", "MSFT", "TSLA", "WMT", "PG", "CVS", "SBUX", "V", "BABA", "KO", "XOM",
        "SPY", "TSM", "ACN", "CSCO", "FB", "PYPL", "NFLX", "VZ", "PG", "MA", "KO", "LLY",
        "PYPL", "DIS", "VEEV", "SBUX", "MMM", "LMT", "PFE", "NKE", "VZ", "COST", "NDAQ", "MS",
        "GM", "ATVI", "MS", "LMT", "TGT", "PLD", "WBA", "GOOG", "BABA", "CSX", "EA", "COP",
        "PFE", "BNSF", "PG", "HCA", "COF", "AXP", "DELL", "T", "TMO", "UNH", "BMY", "AZN",
        "HUM", "CVX", "NSC", "BIDU", "FCX", "GE", "HAL", "K", "MCK", "DD", "AVGO", "GS", "KHC",
        "KO", "MMM", "MCD", "SBUX", "BA", "CAT", "TGT", "WFC", "MSFT", "FDX", "AIG", "V",
        "IBB", "XLF", "SPY", "IWM", "EEM", "VWO", "SCHD", "MSCI", "VOO", "ACWI", "VTI", "F",
        "PYPL", "PFE", "DIS", "UPS", "TGT", "WMT", "FDX", "NEE", "UNP", "HD", "COST", "SWK",
        "NKE", "SCHW", "BIIB", "VLO", "EXC", "DD", "PH", "UNH", "HUM", "NVDA", "AVGO", "COF",
        "AAPL", "META", "MSFT", "GOOG", "XOM", "CVX", "SPY", "AMZN", "NFLX", "TGT", "TSLA",
        "AIG", "AMT", "IBM", "GE", "STZ", "GILD", "PFE", "AMD", "INTC", "BA", "MMM", "F", "SBUX"
    ]


# Function to recommend stocks based on user's selected risk level
def recommend_stocks(risk_level):
    all_stocks = get_all_stocks()  # Retrieve all available stock tickers
    stock_data = {}  # Dictionary to store stock information
    
    for ticker in all_stocks:
        try:
            info = yf.Ticker(ticker).info  # Fetch stock information
            stock_data[ticker] = info  # Store info in dictionary
        except:
            continue  # Skip stocks that cause errors
    
    # Filter stocks based on beta value (a measure of volatility)
    if risk_level == "Low":
        return [ticker for ticker, data in stock_data.items() if data.get("beta", 1) < 1]  # Less volatile stocks
    elif risk_level == "Medium":
        return [ticker for ticker, data in stock_data.items() if 1 <= data.get("beta", 1) <= 1.5]  # Moderate volatility
    elif risk_level == "High":
        return [ticker for ticker, data in stock_data.items() if data.get("beta", 1) > 1.5]  # High volatility stocks
    else:
        print("Invalid risk level. Please choose Low, Medium, or High.")
        exit()

# Get the recommended stocks based on risk level
selected_stocks = recommend_stocks(risk_level)

# Function to retrieve stock price history
def get_stock_data(ticker):
    try:
        stock = yf.Ticker(ticker.strip())  # Fetch stock object
        data = stock.history(start=start_date, end=end_date)  # Retrieve historical price data
        return data['Close'] if not data.empty else None  # Return closing prices if data is available
    except:
        return None  # Return None if fetching fails

# Fetch stock data for the recommended stocks
data = {ticker: get_stock_data(ticker) for ticker in selected_stocks if get_stock_data(ticker) is not None}
stock_df = pd.DataFrame(data)  # Convert to a DataFrame

# If no valid stock data is retrieved, exit the program
if stock_df.empty:
    print("No valid stock data retrieved. Please check your tickers and try again.")
    exit()

# Compute daily returns of the stocks
daily_returns = stock_df.pct_change().dropna()

# Compute stock volatility (annualized standard deviation of returns)
volatility = daily_returns.std() * np.sqrt(252)

# Compute Sharpe Ratio (risk-adjusted return, assuming risk-free rate = 0)
sharpe_ratios = daily_returns.mean() / daily_returns.std() * np.sqrt(252)

# Compute Price-to-Earnings (P/E) ratios
pe_ratios = {}
for ticker in selected_stocks:
    try:
        pe_ratios[ticker] = yf.Ticker(ticker.strip()).info.get('trailingPE', 'N/A')  # Fetch P/E ratio
    except:
        pe_ratios[ticker] = 'N/A'  # If unavailable, return 'N/A'

# Display results in a more aesthetically pleasing way
print("\nRecommended Stocks:")
print("\n".join(selected_stocks))  # List of selected stocks

print("\nStock Data Summary:")
summary_df = pd.DataFrame({
    "Volatility (Std. Dev.)": volatility,
    "Sharpe Ratio": sharpe_ratios,
    "P/E Ratio": pe_ratios
})
print(summary_df)

# Plot volatility and Sharpe ratio for selected stocks
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot Volatility
volatility.sort_values().plot(kind='bar', ax=axes[0], color='salmon', title="Stock Volatility (Annualized)", xlabel="Stocks", ylabel="Volatility")
axes[0].tick_params(axis="x", rotation=45)

# Plot Sharpe Ratios
sharpe_ratios_series = pd.Series(sharpe_ratios).sort_values()
sharpe_ratios_series.plot(kind='bar', ax=axes[1], color='lightgreen', title="Stock Sharpe Ratios", xlabel="Stocks", ylabel="Sharpe Ratio")
axes[1].tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import unittest
from unittest.mock import patch

# Function to fetch all available stock tickers from Yahoo Finance
def get_all_stocks():
    try:
        tickers = yf.Tickers(" ")  # Fetches multiple tickers from Yahoo Finance
        return list(tickers.tickers.keys())  # Returns a list of available stock symbols
    except:
        return []  # Returns an empty list if fetching fails

# Function to recommend stocks based on user's selected risk level
def recommend_stocks(risk_level):
    all_stocks = get_all_stocks()  # Retrieve all available stock tickers
    stock_data = {}  # Dictionary to store stock information
    
    for ticker in all_stocks:
        try:
            info = yf.Ticker(ticker).info  # Fetch stock information
            stock_data[ticker] = info  # Store info in dictionary
        except:
            continue  # Skip stocks that cause errors
    
    # Filter stocks based on beta value (a measure of volatility)
    if risk_level == "Low":
        return [ticker for ticker, data in stock_data.items() if data.get("beta", 1) < 1]  # Less volatile stocks
    elif risk_level == "Medium":
        return [ticker for ticker, data in stock_data.items() if 1 <= data.get("beta", 1) <= 1.5]  # Moderate volatility
    elif risk_level == "High":
        return [ticker for ticker, data in stock_data.items() if data.get("beta", 1) > 1.5]  # High volatility stocks
    else:
        raise ValueError("Invalid risk level. Please choose Low, Medium, or High.")

# Function to retrieve stock price history
def get_stock_data(ticker, start_date, end_date):
    try:
        stock = yf.Ticker(ticker.strip())  # Fetch stock object
        data = stock.history(start=start_date, end=end_date)  # Retrieve historical price data
        return data['Close'] if not data.empty else None  # Return closing prices if data is available
    except:
        return None  # Return None if fetching fails

# Unit Testing Class
class TestStockAnalysis(unittest.TestCase):

    @patch('yfinance.Tickers')  # Mock the yfinance.Tickers to simulate fetching tickers
    def test_get_all_stocks(self, mock_tickers):
        # Simulating that yf.Tickers returns a mocked list of tickers
        mock_tickers.return_value.tickers = {'AAPL': 'Apple', 'GOOGL': 'Alphabet'}
        stocks = get_all_stocks()
        self.assertIsInstance(stocks, list)  # Should return a list
        self.assertIn('AAPL', stocks)  # Ensure 'AAPL' is in the list of tickers
        self.assertIn('GOOGL', stocks)  # Ensure 'GOOGL' is in the list of tickers

    @patch('yfinance.Ticker')  # Mock the yfinance.Ticker to simulate stock data retrieval
    def test_get_stock_data(self, mock_ticker):
        # Simulate fetching valid stock data for AAPL
        mock_ticker.return_value.history.return_value = pd.DataFrame({
            'Close': [150, 152, 153, 155],
            'Date': pd.date_range(start="2023-01-01", periods=4, freq='D')
        })
        
        data = get_stock_data("AAPL", "2023-01-01", "2023-01-04")
        self.assertIsInstance(data, pd.Series)  # Should return a Pandas Series with closing prices
        self.assertEqual(data[0], 150)  # Check the first closing price

        # Simulate no data returned
        mock_ticker.return_value.history.return_value = pd.DataFrame()
        data = get_stock_data("AAPL", "2023-01-01", "2023-01-04")
        self.assertIsNone(data)  # Should return None if no data is fetched

    @patch('yfinance.Ticker')  # Mocking the Ticker to include beta data for stock recommendation
    def test_recommend_stocks(self, mock_ticker):
        # Simulate that each stock has beta values
        mock_ticker.return_value.info = {'beta': 0.8}  # Simulate a low volatility stock for AAPL
        mock_ticker.return_value.history.return_value = pd.DataFrame()  # Mock no price data for simplicity
        
        # Simulate the list of tickers
        with patch('yfinance.Tickers') as mock_tickers:
            mock_tickers.return_value.tickers = {'AAPL': 'Apple', 'GOOGL': 'Alphabet'}
            stocks_low = recommend_stocks("Low")  # Test for low-risk stocks
            self.assertGreater(len(stocks_low), 0)  # Should recommend at least one stock

            # Test for medium-risk stocks
            mock_ticker.return_value.info = {'beta': 1.2}  # Simulate moderate volatility for GOOGL
            stocks_medium = recommend_stocks("Medium")
            self.assertGreater(len(stocks_medium), 0)  # Should recommend at least one stock

            # Test for high-risk stocks
            mock_ticker.return_value.info = {'beta': 1.8}  # Simulate high volatility for GOOGL
            stocks_high = recommend_stocks("High")
            self.assertGreater(len(stocks_high), 0)  # Should recommend at least one stock

# Run tests in Jupyter without exiting
suite = unittest.TestLoader().loadTestsFromTestCase(TestStockAnalysis)
unittest.TextTestRunner().run(suite)
```
